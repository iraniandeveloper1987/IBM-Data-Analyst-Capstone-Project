**Download and Load the Data**


To start, download and load the dataset into a `pandas` DataFrame.


```python
# Step 2: Import necessary libraries and load the dataset
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("cleaned_survey_data_V2.csv")

# Display the first few rows to understand the structure of the data
df.head()
```

### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.


```python
# Extract the 'DatabaseWantToWorkWith' column and split the values
databases = df['DatabaseWantToWorkWith'].dropna().str.split(';').explode()

# Get the top 5 most desired databases
top_databases = databases.value_counts().nlargest(5)

# Plot the pie chart
plt.figure(figsize=(10, 7))
top_databases.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Top 5 Databases Respondents Want to Work With')
plt.ylabel('')  # Hide the y-label

# Add legend
plt.legend(top_databases.index, title="Databases", loc="center left", bbox_to_anchor=(1, 0.5))

plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.


```python
##Write your code here
# Extract the 'DevType' column and split the values
dev_types = df['DevType'].dropna().str.split(';').explode()

# Get the top 5 developer roles
top_dev_types = dev_types.value_counts().nlargest(5)

# Plot the pie chart
plt.figure(figsize=(10, 7))
top_dev_types.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Top 5 Developer Roles')
plt.ylabel('')  # Hide the y-label

# Add legend
plt.legend(top_dev_types.index, title="Developer Roles", loc="best", bbox_to_anchor=(1, 0.5))

plt.show()
```
